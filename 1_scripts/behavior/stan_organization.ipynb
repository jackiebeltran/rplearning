```python
import numpy as np
import pandas as pd  
import glob
import os
```

```python
flist = sorted(glob.glob('/Users/jackiebeltran/Documents/GitHub/RP_Learning/0_data/N_75/*.csv'))
print('participants in this dataset are: ' + str(len(flist)))
```

```python
flist
```

```python
# empty list to crreate a single dataframe with all subjects parameters
df1_dataframes = []

# load and append each df1 dataframe to the list
for file in flist:
    
    df = pd.read_csv(file)
    
    df1_dataframes.append(df)

combined_df = pd.concat(df1_dataframes, ignore_index=True)
```

```python
combined_df
```

```python
def grouping(df):
    def categorize(sub_ID):
        if str(sub_ID).startswith('5'):
            return 'HC'
        elif str(sub_ID).startswith('4'):
            return 'MDD'
        else:
            return 'Other'

    df['group'] = df['sub_ID'].apply(categorize)
    return df

# Apply the function to the DataFrame
df = grouping(combined_df)

df = df.drop(['Unnamed: 0'], axis=1)
```

```python
# save df as csv

df.to_csv('/Users/jackiebeltran/Documents/GitHub/RP_Learning/0_data/stan/data_N75.csv', index=False)
```
